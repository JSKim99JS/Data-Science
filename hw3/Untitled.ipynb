```python
import matplotlib.pyplot as plt
import mplcursors
from datetime import datetime

# Your data
date = [20200501, 20200502, 20200503, 20200504, 20200505]
australia = [0, 0, 0, 0, 0]
china = [27, 27, 44, 44, 59]
japan = [0 ,0 ,0 ,0 ,0]
new_zealand = [0 ,0 ,0 ,0 ,0]
singapore = [0 ,0 ,0 ,0 ,0]
south_korea = [0 ,0 ,0 ,1 ,1]

# Convert date to datetime object
date = [datetime.strptime(str(d),"%Y%m%d") for d in date]

fig = plt.figure(figsize=(15,10))

plt.plot(date,china,label='China',color='red')
plt.plot(date,japan,label='Japan',color='blue')
plt.plot(date,new_zealand,label='New Zealand',color='green')
plt.plot(date,singapore,label='Singapore',color='purple')
plt.plot(date,south_korea,label='South Korea',color='orange')

plt.xlabel('Date')
plt.ylabel('Number of cases')
plt.title('COVID-19 Cases')

# Set xticks
xticks_date = date[::5]
xticks_label = [d.strftime('%Y-%m') for d in date][::5]
plt.xticks(xticks_date,xticks_label)



# Add annotations
lines = plt.gca().get_lines()
cursor = mplcursors.cursor(lines)
@cursor.connect("add")
def on_add(sel):
    x,y,_ = sel.artist[sel.target.index].get_data()
    sel.annotation.set(text=f'{x.strftime("%Y-%m-%d")}\n{sel.artist.get_label()}\n{y}')

plt.legend()
plt.show()
```
